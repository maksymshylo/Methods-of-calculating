```python
import numpy as np
h = 0.01
```

```python
import matplotlib.pyplot as plt
x = list(np.arange(-np.pi,np.pi,0.01))
qwerty = x*np.sin(x)
plt.plot(x,qwerty,"g",label = 'original')
plt.legend(loc='upper left')
plt.show()
```

```python
def f(x, y):
    return ((1 - x ** 2) * (y- x*np.sin(x)) + np.sin(x) +x*np.cos(x))
y = [0]
for i in range(0,len(x)-1):
    k_1 = h * f(x[i], y[i])
    k_2 = h * f(x[i] + h / 2, y[i] + k_1 / 2)
    k_3 = h * f(x[i] + h / 2, y[i] + k_2 / 2)
    k_4 = h * f(x[i] + h,     y[i] + k_3)
    y.append(y[i] + (k_1 + 2 * k_2 + 2 * k_3 + k_4) / 6)
    
plt.plot(x,y,"brown", label = "RK4")
plt.legend(loc='upper left')
plt.show()
```

```python
z = y[:4]
for i in range(3, len(x)-1):
    z.append(z[i]+h*(55 * f(x[i],z[i])-59*f(x[i - 1],z[i - 1])+37*f(x[i - 2],z[i - 2])-9*f(x[i - 3],z[i - 3]))/24)
plt.plot(x,z,"yellow",label = "Adams_Bashfort")
plt.legend(loc='upper left')
plt.show()
```

```python

plt.plot(x,y - qwerty,"b",label = "Error RK4")
plt.plot(x,z - qwerty,"r", label = "Error Adams_Bashfort")
plt.legend(loc='upper left')
plt.show()
print("Max Error RK4",abs(y-qwerty).max())
print("Max Error Adams_Bashfort",abs(z-qwerty).max())
```
